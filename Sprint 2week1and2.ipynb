```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df1=pd.read_csv("steps.csv", sep = ";")
df2=pd.read_csv("survey.csv")

remove_nan = df1.fillna(0)
remove_nan.head()
# with this I remove NaN in df1
```

```python
remove_nan2= df2.fillna(0)
remove_nan2.head()
# with this I remove NaN in df2
```

```python
df4=pd.merge(remove_nan, remove_nan2, on="id")
df4.head()
# with this I merge the two datasets into one, next time I will use df3.
```

```python
df4['mean'] = df4.loc[:, '20-6-2013':'13-5-2014'].mean(axis = 1)

df4.head()
#with this I create the mean column
```

```python
def cut_off(x):
    if x > 25000:
        return float('25000')
    else:
        return x
df4['mean'] = df4['mean'].apply(cut_off)
def below0(x):
    if x < 0:
        return float('nan')
    else:
        return x
df4['mean'] = df4['mean'].apply(below0)
# with this I remove outliers of mean
```

```python
df4['weight'].value_counts()
def cut_off(x):
    if x > 300 or x < 40 : 
        return float('nan')
    else: 
        return x
df4['weight'] = df4['weight'].apply(cut_off)
# with this I remove outliers of weight
```

```python
df4['height'].value_counts()
def cut_off(x):
    if x > 300 or x < 50 : 
        return float('nan')
    else: 
        return x
df4['height'] = df4['height'].apply(cut_off)
# with this I remove outliers of height
```

```python
# I didn't came furter then this, unfortunately. Only if I used mean twice I didn't got a error. If I did this right I also got the rest of the needed results.
x = df4[['mean']] 
y = df4['mean']
lm = LinearRegression()
lm = lm.fit(x, y) 
```

```python
df4['mean_p'] = lm.predict(x) 
df4.head(10)
```

```python
from sklearn.metrics import r2_score 
r2_score(df4['mean'], df4['mean_p'])
```

```python
from sklearn.metrics import mean_squared_error 
from math import sqrt 

sqrt(mean_squared_error(df4['mean'], df4['mean_p']))
```
